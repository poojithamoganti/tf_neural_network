```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import random
import math
import matplotlib.pyplot as plt

t=[]
x1_1_list =[]
x2_1_list =[]
x3_1_list =[]
x_list =[[0 for x in range(3)]for y in range(100)]

#creating 50 samples for first data set
for i in range(50):
  x1_1 = random.uniform(-0.5,0.5)
  x2_1 = random.uniform(-0.5,0.5)
  x3_1= 1
  x1_1_list.append(x1_1)
  x2_1_list.append(x2_1)
  x3_1_list.append(x3_1)
  #print(x1_1_list)
  #print(x2_1_list)
  #print(x3_1_list)
  #print( )
  t.append(0)

#creating 50 samples for second data set
#loop to create 25 negative datasets
for i in range(25):
  r = random.uniform(9,16)
  x1_2 = random.uniform(-3,3)
  x2_2 = - math.sqrt(r - (x1_2)**2)
  x3_2 = 1
  x1_1_list.append(x1_2)
  x2_1_list.append(x2_2)
  x3_1_list.append(x3_2)
  t.append(1)
 
 #loop to create 25 positive datasets
for i in range(25):
  r = random.uniform(9,16)
  x1_2 = random.uniform(-3,3)
  x2_2 = math.sqrt(r - (x1_2)**2)
  x3_2 = 1
  x1_1_list.append(x1_2)
  x2_1_list.append(x2_2)
  x3_1_list.append(x3_2)
  t.append(1)
 
 #adding all x1 ,x2 ,x3 values to a single list  
for i in range(len(x1_1_list)):
    x_list[i][0] = x1_1_list[i]
    x_list[i][1] = x2_1_list[i]
    x_list[i][2] = x3_1_list[i]
 
print ("x",x_list)
print("t" , len(t))

#plotting initial dataset
fig = plt.figure(figsize= (14,14))
ax = fig.add_subplot(1,3,1,projection = '3d')
plt.title('Initial Dataset')
for i in range(100):
  if(t[i]==0):
    ax.scatter( x1_1_list[i] ,x2_1_list[i] ,x3_1_list[i] ,marker = "o" , color = 'blue')
  else:
    ax.scatter(x1_1_list[i] ,x2_1_list[i] ,x3_1_list[i], marker = "s" ,color = 'green')

plt.show()


```

**Sigmoid activation function**

```python
model = keras.Sequential([
        keras.layers.Dense(units = 4 ,input_shape =[3] , activation = tf.nn.sigmoid),
        keras.layers.Dense( units = 5  ,activation = tf.nn.sigmoid),
        keras.layers.Dense( units = 1  ,activation = tf.nn.sigmoid)                   
                          
]) 

model.compile(optimizer ='adam' , loss = 'mean_squared_error')



history = model.fit(x_list, t, epochs =500)
loss = history.history['loss']

epochs = range(len(loss))

plt.plot(epochs, loss, 'r', label='Error')

plt.title('Loss Curve')
plt.legend(loc=0)
plt.figure()
plt.show()
```

```python
model = keras.Sequential([
        keras.layers.Dense(units = 4 ,input_shape =[3] , activation = tf.nn.relu),
        keras.layers.Dense( units = 5  ,activation = tf.nn.relu),
        keras.layers.Dense( units = 1  ,activation = tf.nn.relu)                   
                          
]) 

model.compile(optimizer ='adam' , loss = 'mean_squared_error')



history = model.fit(x_list, t, epochs =500)
loss = history.history['loss']

epochs = range(len(loss))

plt.plot(epochs, loss, 'r', label='Error')

plt.title('Loss Curve')
plt.legend(loc=0)
plt.figure()
plt.show()
```

**When tried with different number of neuron matrix , same relu activation function and adam optimizer we can clearly see there is a slight difference in error graph at first 100 epochs**
